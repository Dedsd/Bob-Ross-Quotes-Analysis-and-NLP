## Import of libs

```python
import json
import random
import spacy
from spacy.lang.en import English 
```

## Loading modules and preparing the database

```python
nlp = spacy.load("en_core_web_sm")
nlp
```

```python
with open('databases/d.json', 'r') as f:
    ross = json.load(f)

print(ross)
```

```python
quotes = nlp(str(ross))
quotes
```

## Systematization

## Finding words and visualizing

```python
from spacy.matcher import PhraseMatcher
from IPython.core.display import HTML
```

```python
matcher = PhraseMatcher(nlp.vocab)
stri = 'happy'
tok = nlp(stri)
matcher.add('Search', None, tok)
text = ""
num = 40
matches = matcher(quotes)

display(HTML(f'<h1>{stri}</h1>'))
display(HTML(f"""<p><strong>Results: </strong>{len(matches)}</p>"""))

for i in matches:
    b = i[1] - num
    if b < 0:
        b = 0
    quotes2 = str(quotes)
    text += str(quotes[b:i[2] + num]).replace(stri, f"<mark>{stri}</mark>")
    text += "<en /><en />"
display(HTML(f"{text}"))
```

## Word cloud

```python
from spacy.lang.en.stop_words import STOP_WORDS
```

```python
print(STOP_WORDS)
```

```python
list = []

for token in quotes:
    list.append(token.text)
```

```python
print(list)
```

```python
list_wstopw = []

for word in list:
    if nlp.vocab[word].is_stop == False:
        list_wstopw.append(word)
```

```python
print(list_wstopw)
```

```python
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt
```

```python
wordcloud = WordCloud(max_font_size=50, max_words=100, background_color="white").generate(''.join(list_wstopw))
plt.figure(figsize=(21,21))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
